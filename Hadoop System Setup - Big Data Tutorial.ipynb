# What Is Scaling In Scaling Out- Big Data Tutorial 5

### How to solve Big Data challenges? Big Data: Data with high volume, variety & velocity

### 2 procedure to solve Big Data challenge

#### 1. Scale-Up

- Increas2 the capacity each time there is need by replacing existing system
   - Year 1: 10GB HD & 2GB RAM 
   - Year 2: 30GB HD & 5G RAM (replaced)
   - Year 3: 100GB HD & 10GB RAM (replaced)
- replacing existing sytem with new each time (year) there is a overhead. 
- Migrate dataset frm 1 disk to another at the same time trying to accomodate all the requirement & dataset. 
- this is not feasible in a real time env where each time we need to increase the volume

#### 2. Scale-Out

- Configure new system when required.
  - Year 1: 10GB HD & 2GB RAM 
  - Year 2: 20GB HD & 3G RAM (newly added disk) 
- add new system to existing system itself
- avoid/reduce overhead of data migration
- if requirement changes, there is no need to replace the entire volume with a new volume, just add more disks
- all system will be able to communicate with each other
- Scale-out better and more convenient than scale-up

Hadoop Distributed File System (HDFS) is built/designed based on Scale-Out approach.

Google File System (GFS): published by Google says that instead of Scale-Up, with Scale-Out approach we can solve huge problems.








# What Are The Problems IN Scaling Out- Big Data Tutorial 6

1. Overhead in File Transfer: 
  -  when you process a data, it brings a file from DB to Server
  - then process the data in the server, bring latency hence overhead
  - Solution: 
      - if I can process the available data there & then in that system itself - there is no latency or NW bandwidth consumed while processing this data
      - remove overhead & build better system
      - create a system which can store multiple copy of aa file into a multiple system
      - even if 1 of the system crash or unavailable, there is no severe impact since the file is available in another system
  - Even better solution:
      - create a system that is able to manage & maintain a metadata in a streamlined basis or able to maintain metainfo / directory structure / file system information which will be capable ebough to update itself on a timely basis.
      - this approach reduces lots of overheads while processing huge volume of data.
      
HDFS: 
 - its a filesystem which allows us to process data in a system where data is already available
 - always supports distributed computation
 - processes what kind of data, amount (vol), which services and processing is going on in other system, and based on that it can assigna job.

# Introduction To Hadoop - Big Data Tutorial 7

1. Based on GFS: Doug Cutting designed a file system called HDFS. HADOOP is the name of a toy elephant belonged to Doug's child

   - HDFS is NOT a database. It's a FILESYSTEM used for Storage.
   - HDFS is a storage unit, NOT a processing unit.
   - Processing unit sits in top of HDFS - ie: MAP REDUCE
   - HADOOP has it's own FILESYSTEM which supports Distributed Computation in a local machine itself
   - it can manage all metainfo by itself
   - Its's designed based on GFS - Google FILESYSTEM

2. Hadoop is an open-source library. Anyone can use/distribute
   - Hadoop is builf on Java JVM that supports large dataset processin in a distributed computation env
   
3. Reliable & Scalable platform
   - Reliability: Hadoop tried to create multiple copies of dataset at any point of time
   - if 1 system crashes, other systems will have backup that can be extracted and the information /dataset can be processed 
   
   - Scalability: if there is a need to add more hardware to be used for this computation in a distributed manner - can be done easily in a couple of minute
   
4. Hadoop can run n num of application at any point in time
   -  we can add any num of machine inside a distributed computation, or any number of machines can be apart of a single Filesystem which creates a cluster - in a sync
   - this cluster will know whats going on in another system with the help of NameNode service, which store metainfo
   
5. MapReduce: a computation framework responsible of execution of data that is available in a distributed env

6. HW Failure: Hadoop is capable to manage & maintain a HW failure. Data will not be lost.

7. Co-ordination between nodes: able to co-ordinate between nodes or machines which are a part of cluster
  - it can assign task / store a dataset inside in Hadoop system


# Components Of Hadoop 1.X - Big Data Tutorial 8

1. **Namenode**: contains metainfo abt each & every file in HDFS env
  - during processing, Namenode is the first point of contact
  - system goes to namenode and ask whether this file is available within my HDFS? if not, I can't execute a job..
  - if the file is not available, namenode allows to store data in that dir.
  - Namenode is an important concept in master-slave architecture
  - Namenode stores info file in form of FSimage (File System Image)
  - it never store data, it **namenode stores only metadata**
  - namenode contains in-memory mapping of all blocks of data

2. **Secondary Namenode**: a node/service that always try to help a namenode to create backup or new meta files
  - is not replacement namenode
  - responsible to create FSimage (we keep storing new info in HDFS and execute new tasks via mapreduce etc)
  - everytime metainfo is generated, namenode is the busiest service in this ecosystem
  - new metainfo will be stored inside a secondary namenode
